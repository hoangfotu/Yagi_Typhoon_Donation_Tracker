```python
def extract_all_tables_to_one_csv(docx_file, output_csv_file):
    # Open the docx file
    doc = Document(docx_file)

    # Initialize a list to hold all tables data
    all_tables = []

    # Loop through all tables
    for table in doc.tables:
        table_data = []
        # Iterate over each row in the table
        for row in table.rows:
            # Extract cell values from the row
            csv_row = [cell.text.strip() for cell in row.cells]
            table_data.append(csv_row)
        # Add table data to all_tables
        all_tables.extend(table_data)
        # Add a blank line between tables for separation (optional)
        all_tables.append([])

    # Convert list of lists to DataFrame
    df = pd.DataFrame(all_tables)

    # Save DataFrame to CSV file
    df.to_csv(output_csv_file, index=False, header=False, encoding='utf-8')

    print(f"All tables extracted and saved as {output_csv_file}")

# Specify the .docx file and the output .csv file
docx_file = '/content/drive/MyDrive/Colab Notebooks/12k Sao Ke/Vietinbank-10-12.docx'
output_csv_prefix = 'df_new.csv'

extract_all_tables_to_one_csv(docx_file, output_csv_prefix)
```

```python
def fix_alignment(df):
    # Create new columns to fix alignment
    df['Fixed_Amount'] = df['Amount'].where(df['Unnamed: 5'].isna(), df['Unnamed: 5'])
    df['Fixed_Donator'] = df['Donator'].where(df['Unnamed: 5'].isna(), np.nan)

    # Fill in the Amount and Donator columns from Fixed columns where necessary
    df['Amount'] = df.apply(lambda row: row['Fixed_Amount'] if pd.notna(row['Fixed_Amount']) else row['Amount'], axis=1)
    df['Donator'] = df.apply(lambda row: row['Fixed_Donator'] if pd.notna(row['Fixed_Donator']) else row['Donator'], axis=1)

    # Drop the temporary columns used for fixing alignment
    df = df.drop(columns=['Fixed_Amount', 'Fixed_Donator'])

    return df
```

```python
def extract_date(date_str):
    # Split by space and loop to find the part that looks like a date (dd/mm/yyyy)
    for part in date_str.split():
        if '/' in part and len(part) == 10:  # Check for a part that contains '/'
            return part
    return date_str  # If no match, return the original string

# Apply the extract_date function to the 'Date' column
df2['Date'] = df2['Date'].apply(extract_date)
```
