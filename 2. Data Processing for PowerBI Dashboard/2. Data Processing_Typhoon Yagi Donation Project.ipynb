Seems to be all 4 df have some case like this:
- First word begin with a series > 10 characters, if it has the last ( . or : or ; ) ---> Take the last characters after that symbal and words follow up that < 10 characters
- If not, take all words having characters < 10
- If is null, then keep original

```python
import re

def process_note(note):
    # Check if there's a period, colon, or semicolon in the first word
    first_word = note.split()[0]
    if '.' in first_word or ':' in first_word or ';' in first_word:
        # Find the position of the last occurrence of period, colon, or semicolon
        last_period = first_word.rfind('.')
        last_colon = first_word.rfind(':')
        last_semicolon = first_word.rfind(';')

        # Determine the last separator's position
        last_sep = max(last_period, last_colon, last_semicolon)

        # Extract the part after the last separator
        after_last_sep = first_word[last_sep + 1:] if last_sep != -1 else first_word

        # Extract words with less than 10 characters from the note
        words = re.findall(r'\b\w{1,9}\b', note)

        # Filter out non-word characters and duplicates while preserving order
        seen = set()
        unique_words = []
        for word in [after_last_sep] + words[1:]:  # Include the part after the last separator
            if word.isalpha() and word not in seen:  # Ensure word is alphabetic and not already added
                seen.add(word)
                unique_words.append(word)

        return ' '.join(unique_words)
    else:
        # If no separator, extract words with less than 10 characters from the note
        words = re.findall(r'\b\w{1,9}\b', note)

        # Filter out non-word characters and duplicates while preserving order
        seen = set()
        unique_words = []
        for word in words:
            if word.isalpha() and word not in seen:
                seen.add(word)
                unique_words.append(word)

        return ' '.join(unique_words)

# Apply the function to the DataFrame
df1['Messages'] = df1['Note'].apply(process_note)
df2['Messages'] = df2['Note'].apply(process_note)
df3['Messages'] = df3['Note'].apply(process_note)
df4['Messages'] = df4['Note'].apply(process_note)
```

#2-Create Donator names

I realize some principle below for Names:
- Can be consecutively CAPITALIZED all characters up to 8 words, at any position
- Same logic, but Capitalized first letters
- Else, keep original


After doing a long time, I realize need to skip misleading names like "Ban Mat Tran To Quoc Viet Nam" or other form. Its the donation fund, not the name of the donators

```python
def extract_name(message):
    # Split the message into words
    words = message.split()

    # Define a list of names/phrases to skip
    skip_names = [
        "Ban Mat Tran To Quoc Viet Nam",
        "Ban Mat Tran To Quoc VN",
        "Mat Tran To Quoc Viet Nam",
        "Ban Cuu Tro TW",
        "Mat Tran To Quoc",
        "Cuu Tro"
    ]

    # Combine skip_names into a lowercased set for easier matching
    skip_names_lower = {name.lower() for name in skip_names}

    # Function to check if a word is capitalized
    def is_capitalized(word):
        return word[0].isupper()

    # Check for sequences where each word starts with a capital letter (left-most priority)
    capitalized_sequence = []
    for word in words:
        if is_capitalized(word):
            capitalized_sequence.append(word)
        else:
            if len(capitalized_sequence) > 1:  # Only consider if we have more than 1 capitalized word
                candidate = ' '.join(capitalized_sequence)
                candidate_lower = ' '.join(capitalized_sequence).lower()

                # Skip if the candidate matches any of the phrases in the skip list
                if candidate_lower not in skip_names_lower:
                    return candidate  # Return the first valid name found
            capitalized_sequence = []  # Reset if non-capitalized word found

    # Check for the last sequence in case the message ends with capitalized words
    if len(capitalized_sequence) > 1:
        candidate = ' '.join(capitalized_sequence)
        candidate_lower = ' '.join(capitalized_sequence).lower()

        if candidate_lower not in skip_names_lower:
            return candidate

    # If no valid name found, return the first 8 words of the message as a fallback
    if len(words) >= 8:
        return ' '.join(words[:8])
    else:
        return ' '.join(words)  # Return all words if fewer than 8

# Apply the function to create the new 'Name' column
df1['Name'] = df1['Messages'].apply(extract_name)
df3['Name'] = df3['Messages'].apply(extract_name)
df4['Name'] = df4['Messages'].apply(extract_name)
```
